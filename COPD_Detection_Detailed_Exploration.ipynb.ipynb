```python
import librosa
import numpy as np
import os
import glob

# Correct folder paths
copd_folder = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/COPD/"
healthy_folder = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/Healthy/"

# Get all .wav files using glob
copd_files = glob.glob(os.path.join(copd_folder, "*.wav"))
healthy_files = glob.glob(os.path.join(healthy_folder, "*.wav"))

# Check if files are found
if not copd_files:
    print("Error: No COPD .wav files found. Check the path.")
if not healthy_files:
    print("Error: No Healthy .wav files found. Check the path.")

# Initialize lists
mfcc_features = []
labels = []

# Function to extract MFCC from an audio file
def extract_mfcc(file_path, num_mfcc=13):
    y, sr = librosa.load(file_path, sr=22050)  # Load audio
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=num_mfcc)  # Extract MFCC
    mfcc_mean = np.mean(mfcc, axis=1)  # Compute mean of each MFCC coefficient
    return mfcc_mean

# Process COPD files
for file_path in copd_files[:26]:  # Select only 26 files
    mfcc_features.append(extract_mfcc(file_path))
    labels.append(1)  # COPD label

# Process Healthy files
for file_path in healthy_files[:26]:  # Select only 26 files
    mfcc_features.append(extract_mfcc(file_path))
    labels.append(0)  # Healthy label

# Convert to NumPy arrays
mfcc_features = np.array(mfcc_features)
labels = np.array(labels)

# Print shapes
print("MFCC Feature Matrix Shape:", mfcc_features.shape)
print("Labels Shape:", labels.shape)
```

```python
def extract_mfcc_features(file_path, sr=22050, n_mfcc=13):
    y, sr = librosa.load(file_path, sr=sr)  # Load audio
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)  # Extract MFCC
    return np.mean(mfcc, axis=1)  # Take mean across time

import os
copd_folder = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/COPD/"
copd_files = [f for f in os.listdir(copd_folder) if f.endswith('.wav')]

if copd_files:
    sample_audio = os.path.join(copd_folder, copd_files[0]) # Use the first audio file
    mfcc_features = extract_mfcc_features(sample_audio)
    print("MFCC Feature Shape:", mfcc_features.shape)
    print("MFCC Features:", mfcc_features)
else:
    print("No .wav files found in the COPD directory. Please check the path.")
```

```python
def extract_wavelet_features(file_path, wavelet='db4', level=5):
    y, sr = librosa.load(file_path, sr=22050)  # Load audio file
    coeffs = pywt.wavedec(y, wavelet, level=level)  # Compute Discrete Wavelet Transform
    features = np.hstack([np.mean(c) for c in coeffs])  # Compute mean of each coefficient
    return features

# Example usage
wavelet_features = extract_wavelet_features(sample_audio)

print("Wavelet Feature Shape:", wavelet_features.shape)
print("Wavelet Features:", wavelet_features)
```

```python
def extract_combined_features(file_path):
    mfcc_features = extract_mfcc_features(file_path)
    wavelet_features = extract_wavelet_features(file_path)

    combined_features = np.hstack((mfcc_features, wavelet_features))  # Merge both features
    return combined_features

# Example usage
combined_features = extract_combined_features(sample_audio)

print("Combined Feature Shape:", combined_features.shape)
print("Combined Features:", combined_features)
```

```python
# Function to extract MFCC features
def extract_mfcc(file_path, n_mfcc=20):
    y, sr = librosa.load(file_path, sr=None)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    mfccs_mean = np.mean(mfccs, axis=1)  # Take mean to get fixed-length feature vector
    return mfccs_mean

# Function to extract Wavelet Transform features
def extract_wavelet(file_path, wavelet='db4', level=3):
    y, sr = librosa.load(file_path, sr=None)
    coeffs = pywt.wavedec(y, wavelet, level=level)
    wavelet_features = np.hstack([np.mean(c) for c in coeffs])  # Taking mean of coefficients
    return wavelet_features

# Function to process all audio files and create dataset
def create_dataset(folder_path, label):
    features = []
    labels = []
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        if file_path.endswith('.wav'):  # Ensure only audio files are processed
            try:
                mfcc_features = extract_mfcc(file_path)
                wavelet_features = extract_wavelet(file_path)
                combined_features = np.hstack((mfcc_features, wavelet_features))
                features.append(combined_features)
                labels.append(label)
            except Exception as e:
                print(f"Error processing {file}: {e}")
    return np.array(features), np.array(labels)

# Define dataset paths
copd_path = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/COPD/"
healthy_path = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/Healthy/"

# Extract features for both classes
X_copd, y_copd = create_dataset(copd_path, label=1)
X_healthy, y_healthy = create_dataset(healthy_path, label=0)

# Combine datasets
X = np.vstack((X_copd, X_healthy))
y = np.hstack((y_copd, y_healthy))

# Print dataset shape
print(f"Feature matrix shape: {X.shape}, Labels shape: {y.shape}")
```

```python
import os
import numpy as np
import librosa
import pywt
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D, MaxPooling1D, Flatten
```

```python
# Function to extract MFCC features
def extract_mfcc(file_path, n_mfcc=20):
    y, sr = librosa.load(file_path, sr=None)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfccs, axis=1)  # Take mean to get fixed-length feature vector

# Function to extract Wavelet Transform features
def extract_wavelet(file_path, wavelet='db4', level=3):
    y, sr = librosa.load(file_path, sr=None)
    coeffs = pywt.wavedec(y, wavelet, level=level)
    return np.hstack([np.mean(c) for c in coeffs])  # Taking mean of coefficients

# Function to create dataset from folder
def create_dataset(folder_path, label):
    features, labels = [], []
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        if file_path.endswith('.wav'):
            try:
                mfcc_features = extract_mfcc(file_path)
                wavelet_features = extract_wavelet(file_path)
                combined_features = np.hstack((mfcc_features, wavelet_features))
                features.append(combined_features)
                labels.append(label)
            except Exception as e:
                print(f"Error processing {file}: {e}")
    return np.array(features), np.array(labels)
```

```python
import os
import numpy as np
import librosa
import pywt
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.utils import to_categorical

# Set the fixed feature size (adjust this according to your requirements)
fixed_feature_size = 104  # Example size after wavelet transform

# Function to extract features with wavelet transform
def extract_features_with_wavelet(file_path):
    y, sr = librosa.load(file_path, sr=None)

    # Extract MFCC
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

    # Apply wavelet transform
    coeffs = pywt.dwt(mfcc, 'db1')
    cA, cD = coeffs
    wavelet_features = np.concatenate((cA.flatten(), cD.flatten()))

    # Ensure the feature vector has the fixed size by padding/truncating
    if len(wavelet_features) < fixed_feature_size:
        # Pad with zeros if the feature length is shorter than the fixed size
        wavelet_features = np.pad(wavelet_features, (0, fixed_feature_size - len(wavelet_features)), 'constant')
    else:
        # Truncate if the feature length is longer than the fixed size
        wavelet_features = wavelet_features[:fixed_feature_size]

    return wavelet_features

# Function to load data
def load_data(copd_folder, healthy_folder):
    features = []
    labels = []

    # COPD files (label = 1)
    for file in os.listdir(copd_folder):
        if file.endswith('.wav'):
            path = os.path.join(copd_folder, file)
            feat = extract_features_with_wavelet(path)
            features.append(feat)
            labels.append(1)

    # Healthy files (label = 0)
    for file in os.listdir(healthy_folder):
        if file.endswith('.wav'):
            path = os.path.join(healthy_folder, file)
            feat = extract_features_with_wavelet(path)
            features.append(feat)
            labels.append(0)

    return np.array(features), np.array(labels)

# Set paths to your folders
copd_folder = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/COPD/"
healthy_folder = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/Healthy/"

# Load features and labels
X, y = load_data(copd_folder, healthy_folder)

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reshape X to fit the CNN model input
X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

# 5-fold cross-validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
fold_no = 1
cnn_accuracies = []

for train_index, val_index in kf.split(X_scaled, y):
    X_train, X_val = X_scaled[train_index], X_scaled[val_index]
    y_train, y_val = y[train_index], y[val_index]

    # Convert labels to categorical (one-hot encoding)
    y_train = to_categorical(y_train, num_classes=2)
    y_val = to_categorical(y_val, num_classes=2)

    # CNN Model
    model = Sequential()
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(X_scaled.shape[1], 1)))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dense(2, activation='softmax'))  # 2 classes: Healthy (0) and COPD (1)

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    model.fit(X_train, y_train, epochs=10, batch_size=2, verbose=0)

    # Evaluate the model
    _, accuracy = model.evaluate(X_val, y_val, verbose=0)
    cnn_accuracies.append(accuracy)
    print(f"🌲 Fold {fold_no} Accuracy: {accuracy:.4f}")

    # Classification Report and Confusion Matrix for each fold
    y_pred = model.predict(X_val)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_val, axis=1)

    print(f"Classification Report for Fold {fold_no}:\n")
    print(classification_report(y_true_classes, y_pred_classes))
    print(f"Confusion Matrix for Fold {fold_no}:\n")
    print(confusion_matrix(y_true_classes, y_pred_classes))

    fold_no += 1

# Mean accuracy for CNN model
mean_accuracy = np.mean(cnn_accuracies)
print(f"\nCNN Mean Accuracy: {mean_accuracy:.4f}")
```

```python
import os
import numpy as np
!pip install pyWavelets
import pywt
import librosa
from sklearn.preprocessing import StandardScaler

# Paths to COPD & Healthy .wav files
copd_folder = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/COPD/"
healthy_folder = "/content/drive/MyDrive/KaggleDatasets/Balanced_Respiratory_Sound_Database/Healthy/"

def extract_rich_features(file_path, n_mfcc=40):
    signal, sr = librosa.load(file_path, sr=None)

    # ============ MFCCs ============
    mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=n_mfcc)
    mfcc_mean = np.mean(mfcc, axis=1)
    mfcc_std = np.std(mfcc, axis=1)

    # Delta and Delta-Delta
    delta = librosa.feature.delta(mfcc)
    delta2 = librosa.feature.delta(mfcc, order=2)
    delta_mean = np.mean(delta, axis=1)
    delta2_mean = np.mean(delta2, axis=1)

    # ============ Spectral Features ============
    spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=signal, sr=sr))
    spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=signal, sr=sr))
    spectral_rolloff = np.mean(librosa.feature.spectral_rolloff(y=signal, sr=sr))
    zero_crossing_rate = np.mean(librosa.feature.zero_crossing_rate(y=signal))  # 🔧 Fixed here

    spectral_features = np.array([
        spectral_centroid, spectral_bandwidth, spectral_rolloff, zero_crossing_rate
    ])

    # ============ Wavelet Features ============
    coeffs = pywt.wavedec(signal, 'db4', level=5)  # 🔧 Use signal instead of undefined 'y'
    wavelet_stats = []
    for c in coeffs:
        wavelet_stats.extend([
            np.mean(c),
            np.std(c),
            np.sum(np.square(c)),  # energy
            -np.sum(c * np.log(np.abs(c) + 1e-6))  # entropy
        ])

    # ============ Combine All ============
    return np.hstack((
        mfcc_mean, mfcc_std,
        delta_mean, delta2_mean,
        spectral_features,
        wavelet_stats
    ))

# Lists to hold features and labels
features = []
labels = []

# Process COPD files (label = 1)
for file in sorted(os.listdir(copd_folder)):
    if file.endswith(".wav"):
        f = extract_rich_features(os.path.join(copd_folder, file))
        features.append(f)
        labels.append(1)

# Process Healthy files (label = 0)
for file in sorted(os.listdir(healthy_folder)):
    if file.endswith(".wav"):
        f = extract_rich_features(os.path.join(healthy_folder, file))
        features.append(f)
        labels.append(0)

# Convert to NumPy arrays
X = np.array(features)
y = np.array(labels)

# Normalize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

print("✅ Feature shape:", X.shape)
print("✅ Labels shape:", y.shape)

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def compute_all_metrics(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()

    accuracy = accuracy_score(y_true, y_pred) * 100
    precision = precision_score(y_true, y_pred) * 100
    recall = recall_score(y_true, y_pred) * 100
    f1 = f1_score(y_true, y_pred) * 100
    specificity = (tn / (tn + fp)) * 100

    return accuracy, precision, recall, f1, specificity
```
